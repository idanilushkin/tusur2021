1) Имеется следующая выборка наблюдений некоторого случайного эксперимента: 
428, 300, 250, 578, 492, 204, 245, 420, 760, 363, 600, 483, 711, 265, 572, 350, 652, 460, 340, 670, 565, 750, 452. 
Каково среднее значение по этой выборке?

```python
import numpy as np
import statistics as stat
A=[428, 300, 250, 578, 492, 204, 245, 420, 760, 363, 600, 483, 711, 265, 572, 350, 652, 460, 340, 670, 565, 750, 452]
print("%.2f" % np.mean(A))
print("%.2f" % stat.mean(A))
```

2) Какое из представленных в выборке значений является ее медианой??

```python
print(stat.median(A))
print(np.median(A))
```

3) Какова величина среднеквадратичного отклонения указанной ранее выборки?

```python
print("%.2f" % stat.pstdev(A))
print("%.2f" % np.std(A))
```

4) Даны две выборки одинаковой длины: X = (428, 300, 250, 578, 492, 204, 245, 420, 760, 363, 600, 483, 711, 265, 572, 350, 652, 460, 340, 670, 565, 750, 452) и Y = (21, 33, 43, 71, 85, 17, 38, 43, 53, 76, 33, 16, 24, 59, 65, 73, 45, 63, 53, 43, 58, 28, 37). Найдите ковариацию cov(X,Y).

```python
X = [428, 300, 250, 578, 492, 204, 245, 420, 760, 363, 600, 483, 711, 265, 572, 350, 652, 460, 340, 670, 565, 750, 452]
Y = [ 21,  33,  43,  71,  85,  17,  38,  43,  53,  76,  33,  16,  24,  59,  65,  73,  45,  63,  53,  43,  58,  28,  37]

mX = X-np.mean(X)
mY = Y-np.mean(Y)
print("%.4f" % np.mean(mX*mY))

covax=np.cov(X,Y, bias=True)
print("%.4f" % covax[1][0])
```

5) Найти коэффициент Пирсона

```python
print("%.4f" % np.corrcoef(X,Y)[1][0])
```
