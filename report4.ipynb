В разделе открытых данных на сайте Министерства здравоохранения https://minzdrav.gov.ru/opendata скачайте данные о "Перечне жизненно необходимых и важнейших лекарственных препаратов для медицинского применения" за все годы, какие там найдете. 

Постройте график, отображающий эту информацию в удобном и наглядном виде, так чтобы можно было сравнить разные годы, какие изменения были, если они были. График должен иметь стандартное оформление: Подписи осей, заголовок, легенду.

```python
import pandas as pd
import numpy as np

#Чтение данных из файла
f16 = pd.read_json('data2016.json')
pd.set_option('display.max_rows', 40) #не работает
f16
```

```python
f16.columns
```

```python
f17 = pd.read_json('data2017.json')
f17
```

```python
CountCode16 = f16['atccode1'].value_counts()
# CC16s = CountCode16.sort_index(inplace=False)
# CC16s
```

```python
CountCode17 = f17['atccode1'].value_counts()
CountCode17
```

```python
import matplotlib.pyplot as plt
# Создание объекта Figure
fig = plt.figure(figsize=(10,5))
plt.plot(CountCode16, color='g', ls='', marker='o',markersize=12, label='2016')
plt.plot(CountCode17, color='c', ls='', marker='*', markersize=8, label='2017')
plt.xlabel('Код анатомо-терапевтическо-химической классификации', fontsize=14) # добавляем подпись к оси абцисс "ось х"
plt.ylabel('Количество препаратов, шт', fontsize=14) # добавляем подпись к оси ординат "ось y"
plt.title('Количество препаратов каждой группы по годам', fontsize=16)
plt.legend(['2016','2017'])
```
